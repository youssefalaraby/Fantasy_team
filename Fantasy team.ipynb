```python
Top_10=gk_df.sort_values(by='Total Points',ascending=False).head(10)
Top_10
```

#### I will analyze each feature with the cost of players

```python
# top 10 GK have Points
plt.figure(figsize=(12,10))
x1=sns.scatterplot(x=Top_10['Clean Sheets'],y=Top_10['Total Points'],s=500,hue=Top_10['Cost Today']);
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+.15, point['y']-.3, str(point['val']))
label_point(Top_10['Clean Sheets'],Top_10['Total Points'],Top_10['name'],x1)
plt.title("Total Points and Clean Sheets of top 10 GK")
```

#### XGI stands for Expected Goals Added. It is a metric used to measure the attacking contribution of a player in football
#### A player with a high XGI is likely to be a good attacking player, as they are creating and finishing chances that are expected to lead to goals.

```python
plt.figure(figsize=(10,10))
sns.barplot(x=top_mid['xGi'].head(15),y=top_mid['name'].head(15))
plt.title('xGi for each player')
```
